## SIMPLE RECOMENDED

```python
vote_data=meta[['original_title','vote_average','vote_count']]
```

```python
vote_data['vote_average'].mean()
```

```python
def weight_rating(row):
    v=row['vote_count'] # v is the number of votes for the movie
    m=1500 # minimal vote count yang ingin di tampilkan di chat 'm is the minimum votes required to be listed in the chart
    r=row['vote_average'] #R is the average rating of the movie
    c=5.6 # mean vote_average C is the mean vote across the whole report
    wr=((v/(v+m)*r)+(m/(v+m)*c))
    return wr
```

```python
vote_data['score'] = vote_data.apply(weight_rating,axis=1)
vote_data
```

```python
vote_data.sort_values(by='score',ascending=False)
```

```python
vote_data.sort_values(by='vote_count',ascending=False)
```

```python
def newgenres(x):
    genres=[]
    x=literal_eval(x)
    for item in x:
        genres.append(item['name'])
    return ' '.join(genres)
```

```python
def new_cast(x):
    casts = []
    x = literal_eval(x)
    for item in x:
        name = item['name'].replace(' ','')
        casts.append(name)
    return ' '.join(casts)
```

```python
cb_genres_cast_crew['new_crew'] = cb_genres_cast_crew['crew'].apply(new_cast)
```

```python
cb_genres_cast_crew['new_cast'] = cb_genres_cast_crew['cast'].apply(new_cast)
```

```python
cv_res_cast = CountVectorizer().fit_transform(cb_genres_cast_crew['new_cast'])
cos_sim_cast = cosine_similarity(cv_res_cast)
```

```python
def get_recommendation_based_cast(title):
    index_to_search = cb_genres_cast_crew[cb_genres_cast_crew['original_title'] == title].index[0]
    series_similar = pd.Series(cos_sim_cast[index_to_search])
    index_similar = series_similar.sort_values(ascending=False).head(10).index
    return cb_genres_cast_crew.loc[index_similar][['original_title','new_cast']]
```

### Recommendation based on director 30 % , genres 40 % , cast 30 %

```python
literal_eval(cb_genres_cast_crew['crew'].iloc[0])
```

```python
def new_director(x):
    
    for item in literal_eval(x):
        if (item['job'] == 'Director'):
            return item['name'].replace(' ','')
```

```python
cb_genres_cast_crew['director'] = cb_genres_cast_crew['crew'].apply(new_director)
```

> Measure the similarity each film by director

```python
cv_director = CountVectorizer().fit_transform(cb_genres_cast_crew['director'])
sim_director = cosine_similarity(cv_director)
```

```python
pd.Series(sim_director[0]).sort_values(ascending=False).head(10)
```

> Combine All Metric for Content Based

```python
sim_combination = (con_sim_movie * 0.5) + (cos_sim_cast * 0.4) + (sim_director * 0.1)
```

```python
def get_recommendation_combination(title):
    index_movie = cb_genres_cast_crew[cb_genres_cast_crew['original_title'] == title].index[0]
    index_similar = pd.Series(sim_combination[index_movie]).sort_values(ascending=False).head(10).index
    return cb_genres_cast_crew[['original_title','new_cast','newgenres','director']].iloc[index_similar]
```

```python
get_recommendation_combination('Toy Story')
```

```python
def get_cf_item_based(movies):
    serr = []
    for item in movies:
        ser_similar = df_item_similarity[item[0]] * (item[1] - 2.5)
        serr.append(ser_similar)
    similar_total = sum(serr)
    return similar_total.sort_values(ascending=False).head(10)
```

```python
get_cf_item_based([['A Perfect Murder',5],['Titanic',5],['Frankenstein',1]])
```

```python
def get_recommendation_user_based(ratings):
    movie_index = np.array(ratings)[:,0]
    movie_rating = np.array(ratings)[:,1]
    user_rating_pivot.loc[(user_rating_pivot.index[-1] + 1) , movie_index] = movie_rating
    user_rating_pivot.fillna(0,inplace=True)
    similarity = cosine_similarity(user_rating_pivot)
    user_id_similar = pd.DataFrame(similarity,columns=user_rating_pivot.index,
                                   index=user_rating_pivot.index).loc[user_rating_pivot.index[-1]].sort_values(ascending=False)[1:11].index
    mov_serr = []
    for id in user_id_similar:
        current_rating = user_rating_pivot.loc[id]
        mov_serr.append(current_rating)
    return (sum(mov_serr) / 10).sort_values(ascending=False)
    
```
